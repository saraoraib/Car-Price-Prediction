```python
def process_passenger_capacity(value):
    try:
        if pd.notna(value):
            return int(str(value).split('+')[0])
        else:
            return np.NaN
    except ValueError:
        return np.NaN
```

```python
data['عدد الركاب'] = data['عدد الركاب'].apply(process_passenger_capacity)
```

```python
base=['موديل سنة','وسادة حماية هوائية','مُكيّف','إغلاق مركزي','جهاز إنذار','فتحة سقف',
     'أصل السيارة', 'أصحاب سابقون', 'لون السيارة', 'الموديل','رخصة السيارة','نوع الوقود','نوع الجير','الزجاج']
```

```python
sorted(data['موديل سنة'].unique())
```

```python
def prepare_x(df, base):
    data = df.copy()
    features = base.copy()
    
    data['age'] = 2022 - data['موديل سنة']
    features.append('age')
    df_num = data[features]
    x = df_num
    return x
```

```python
def one_hot_encode(df, categorical_columns):
    df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
    return df_encoded
```

```python
X = prepare_x(data, base)
y = data['log_price'] 
```

```python
categorical_columns = ['أصل السيارة', 'أصحاب سابقون', 'لون السيارة', 'الموديل','رخصة السيارة','نوع الوقود','نوع الجير','الزجاج']
X_encoded = one_hot_encode(X, categorical_columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Polynomial Regression
poly_model = make_pipeline(PolynomialFeatures(degree=1), LinearRegression())
poly_model.fit(X_train_scaled, y_train)
poly_pred = poly_model.predict(X_test_scaled)
poly_r2 = r2_score(y_test, poly_pred)
```

```python
# K-Nearest Neighbors
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(X_train_scaled, y_train)
knn_pred = knn_model.predict(X_test_scaled)
knn_r2 = r2_score(y_test, knn_pred)
```

```python
# Decision Tree
tree_model = DecisionTreeRegressor(random_state=42)
tree_model.fit(X_train_scaled, y_train)
tree_pred = tree_model.predict(X_test_scaled)
tree_r2 = r2_score(y_test, tree_pred)
```

```python
# Display R-squared values
print(f"Polynomial Regression R-squared: {poly_r2:.4f}")
print(f"K-Nearest Neighbors R-squared: {knn_r2:.4f}")
print(f"Decision Tree R-squared: {tree_r2:.4f}")
```
